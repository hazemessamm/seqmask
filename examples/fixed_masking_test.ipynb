```python
import os
os.environ["SEQMASK_BACKEND"] = "torch"
from seqmask import FixedMasking
import torch
```

```python
masker = FixedMasking(0.15, 0.8, 0.1, 3, list(range(4, 100)), [0, 1, 2, 3])
```

```python
# Dummy test to demonstrate the rates of masking, mutation and leaving tokens as they are.
c_mask = 0 # count mask
c_mut = 0 # count mutations
c_left = 0 # count left
num_iterations = 10000
x = torch.randint(4, 100, (1, 100))
for _ in range(num_iterations):
    a, b = masker(x)
    m = (a == 3).float().mean()
    mut = ((a != x) & (a != 3)).float().mean()
    left = ((a == x) & (b != -100)).float().mean()
    c_mask += m
    c_mut += mut
    c_left += left

avg_mask = c_mask / num_iterations
avg_mutation = c_mut / num_iterations
avg_left_out = c_left / num_iterations
print(avg_mask, avg_mutation, avg_left_out, avg_mask + avg_mutation + avg_left_out) # the sum should be around 0.15 which is the same as MLM prob set above.
```
